```python
import math


class HeapOject(object):
    """
    This object realize the data structure of heap.
    """
    def __init__(self, array):
        """
        """
        self._list = array
        self._list_len = len(self._list)
        
    def heap_size(self, A):
        """
        """
        return len(A)
    
    def i_parent(self, i):
        """
        param: i,
        type: int, i is the index plus 1 of array, begin with 1.
        notice: python array index begin with 0.
        
        return: the value of i parent node.
        """
        if self.heap_size(self._list) < 1:
            raise Exception("array is empty.")
        if i == 1:
            return self._list[0]
        return int(math.floor(i/2))
    
    def i_left(self, i):
        """
        """
        k = 2*i
#         if self.heap_size() < k:
#             raise Exception("array out of index.")
        return k
    
    def i_right(self, i):
        """
        """
        k = 2*i + 1
#         if self.heap_size() < k:
#             raise Exception("array out of index")
        return k
            
    def _max_heapify(self, A, i):
        """
        """
        l = self.i_left(i)
        r = self.i_right(i)
        if l <= self.heap_size(A) and self._list[l-1] > self._list[i-1]:
            largest = l
        else:
            largest = i
        
        if r <= self.heap_size(A) and self._list[r-1] > self._list[largest-1]:
            largest = r
        if largest != i:
            tmp = self._list[i-1]
            self._list[i-1] = self._list[largest-1]
            self._list[largest-1] = tmp
            self._max_heapify(A, largest)
    
    def max_heapify(self, A, i):
        self._max_heapify(A, i)
        
    def build_max_heap(self):
        """
        return the max heap, satisfies A[parent(i)] >= A[i]
        """
        for i in xrange(int(math.floor(
        self.heap_size(self._list)/2)), 0, -1):
            self.max_heapify(self._list, i)
            
    def heap_sort(self):
        """
        return sorted heap array, ordered by ASC.
        """
        self.build_max_heap()
        print self.get_array
        j = 0
        for i in xrange(self.heap_size(self._list), 1, -1):
            tmp = self._list[0] 
            self._list[0] = self._list[i-1]
            self._list[i-1] = tmp
            j += 1 
            self.max_heapify(self._list[:-j], 1)
            
    @property        
    def get_array(self):
        """
        """
        return self._list
            
```
